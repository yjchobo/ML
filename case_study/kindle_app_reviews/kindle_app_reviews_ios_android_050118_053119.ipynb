```python
# data cleaning
# drop
ios_drops = ["Platform","Country","App ID","App Name","Publisher ID","Publisher Name","User"]
android_drops = ["Platform","Language","App ID","App Name","Publisher ID","Publisher Name","User","Title"]
for i in ios_drops:
    df_as_kindle_05_18_05_19 = df_as_kindle_05_18_05_19.drop(columns = i, axis = 1, errors = "ignore")

for i in android_drops:
    df_gp_kindle_05_18_05_19 = df_gp_kindle_05_18_05_19.drop(columns = i, axis = 1, errors = "ignore")

# string to datetime
df_as_kindle_05_18_05_19['Date'] = pd.to_datetime(df_as_kindle_05_18_05_19['Date'])
df_gp_kindle_05_18_05_19['Date'] = pd.to_datetime(df_gp_kindle_05_18_05_19['Date'])

# add week and month level
def time_window(df):
    time_windows = ["year","month","week", "day", "dow"]
    for i in time_windows:
        if i == "year":
            df[i] = df["Date"].dt.year
        elif i == "month":
            df[i] = df["Date"].dt.month
        elif i == "week":
            df[i] = df["Date"].dt.week
        elif i == "day":
            df[i] = df["Date"].dt.day
        else:
            df[i] = df["Date"].dt.dayofweek

#time_window(df_as_kindle_05_18_05_19)
df_as_kindle_05_18_05_19.pipe(time_window)
df_gp_kindle_05_18_05_19.pipe(time_window)
#time_window(df_gp_kindle_05_18_05_19)

# week start and month start
# date to index
df_as_kindle_05_18_05_19.set_index("Date", inplace = True)
df_gp_kindle_05_18_05_19.set_index("Date", inplace = True)

df_as_kindle_05_18_05_19['week_start'] = df_as_kindle_05_18_05_19.index.to_period('W').start_time
df_as_kindle_05_18_05_19['month_start'] = df_as_kindle_05_18_05_19.index.to_period('M').start_time

df_gp_kindle_05_18_05_19['week_start'] = df_gp_kindle_05_18_05_19.index.to_period('W').start_time
df_gp_kindle_05_18_05_19['month_start'] = df_gp_kindle_05_18_05_19.index.to_period('M').start_time

# Day of week mapping change
dow_dict = {0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"}
df_as_kindle_05_18_05_19.replace({"dow": dow_dict}, inplace = True)
df_gp_kindle_05_18_05_19.replace({"dow": dow_dict}, inplace = True)

#dropping records without review texts
df_as_kindle_05_18_05_19.dropna(subset = ['Review'],inplace=True)
df_gp_kindle_05_18_05_19.dropna(subset = ['Review'],inplace=True)
```

```python
#Group and normalize similar words (i.e. sync, syncs, syncing)

def manual_kw_normalizer(df):
    
    #sync
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[s][y][n][c][a-z]*", value = r"sync", regex=True) #preserving whispersync
    #download
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[d][o][w][n][l][o][a][d][a-z]*", value = r"download", regex=True)
    #bookmark
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[b][o][o][k][m][a][r][k][a-z]*", value = r"bookmark", regex=True)
    #bug
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[b][u][g][a-z]*", value = r"bug", regex=True)
    #bookshelf
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[b][o][o][k][s][h][e][l][f][a-z]*", value = r"bookshelf", regex=True)
    #borrow
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[b][o][r][r][o][w][a-z]*", value = r"borrow", regex=True)
    #buy
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[b][u][y][a-z]*", value = r"buy", regex=True)
    #choice
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[c][h][o][i][c][e][a-z]*", value = r"choice", regex=True)
    #collection
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[c][o][l][l][e][c][t][i][o][n][a-z]*", value = r"collection", regex=True)
    #color
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[c][o][l][o][r][a-z]*", value = r"color", regex=True)
    #crash
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[c][r][a][s][h][a-z]*", value = r"crash", regex=True)
    #delete
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[d][e][l][e][t][a-z]*", value = r"delete", regex=True)
    #dictionary
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[d][i][c][t][i][o][n][a][r][a-z]*", value = r"dictionary", regex=True)
    #difficult
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[d][i][f][f][i][c][u][l][t][a-z]*", value = r"difficult", regex=True)
    #experience
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[e][x][p][e][r][i][e][n][c][e][a-z]*", value = r"experience", regex=True)
    #fast
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][a][s][t][a-z]*", value = r"fast", regex=True)
    #feature
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][e][a][t][u][r][e][a-z]*", value = r"feature", regex=True)
    #find
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][i][n][d][a-z]*", value = r"find", regex=True)
    #search
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[s][e][a][r][c][h][a-z]*", value = r"search", regex=True)
    #format
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][o][r][m][a][t][a-z]*", value = r"format", regex=True)
    #freeze
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][r][e][e][z][e][a-z]*", value = r"freeze", regex=True)
    #glitch
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[g][l][i][t][c][h][a-z]*", value = r"glitch", regex=True)
    #genre
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[g][e][n][r][e][a-z]*", value = r"genre", regex=True)
    #frustration
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][r][u][s][t][a-z]*", value = r"frustration", regex=True)
    #highlight
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[h][i][g][h][l][i][g][h][t][a-z]*", value = r"highlight", regex=True)
    #improvement
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[i][m][p][r][o][v][e][a-z]*", value = r"improve", regex=True)
    #library
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[l][i][b][r][a][r][a-z]*", value = r"library", regex=True)
    #notification
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[n][o][t][i][f][i][c][a][t][i][o][n][a-z]*", value = r"notification", regex=True)
    #organize
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[o][r][g][a][n][i][z][a-z]*", value = r"organize", regex=True)
    #pdf
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[p][d][f][a-z]*", value = r"pdf", regex=True)
    #purchase
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[p][u][r][c][h][a][s][a-z]*", value = r"purchase", regex=True)
    #recommendation
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[r][e][c][o][m][m][e][n][d][a][t][i][o][n][a-z]*", value = r"recommendation", regex=True)
    #sample
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[s][a][m][p][l][a-z]*", value = r"sample", regex=True)
    #scroll
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[s][c][r][o][l][l][a-z]*", value = r"scroll", regex=True)
    #suggestion
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[s][u][g][g][e][s][t][i][o][n][a-z]*", value = r"suggestion", regex=True)
    #update
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[u][p][d][a][t][e][a-z]*", value = r"update", regex=True)
    #uninstall
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[u][n][i][n][s][t][a][l][l][a-z]*", value = r"uninstall", regex=True)
    #install
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[i][n][s][t][a][l][l][a-z]*", value = r"install", regex=True)
    #disappear
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[d][i][s][a][p][p][e][a][r][a-z]*", value = r"disappear", regex=True)
    #function
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][u][n][c][t][i][o][n][a-z]*", value = r"function", regex=True)
    #font
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[f][o][n][t][a-z]*", value = r"font", regex=True)
    #option
    df["Review1"] = df['Review1'].replace(to_replace=r"[a-z0-9]*[o][p][t][i][o][n][a-z]*", value = r"option", regex=True)
```
